```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_curve
from sklearn.preprocessing import StandardScaler  # Para normalização (média 0, variância 1)

# Carregar o dataset do ficheiro pickle (ajuste o caminho se necessário)
D = pickle.load(open('pimaDiabetes.p', 'rb'))  # Ou use o nome do ficheiro fornecido no enunciado

X = D['data']  # Features: shape (768, 8)
y = D['target']  # Target: shape (768,) - 0: sem diabetes, 1: com diabetes

# Verificar dimensões e balanço das classes
print(f"X.shape: {X.shape}")
print(f"y.shape: {y.shape}")
print(f"Número de exemplos positivos (diabetes): {np.sum(y)}")  # Deve ser ~268
print(f"Número de exemplos negativos (sem diabetes): {np.sum(y == 0)}")  # Deve ser ~500

# Visualizar distribuição das classes (histograma)
plt.figure(figsize=(6, 4))
plt.hist(y, bins=2, edgecolor='black')
plt.xlabel('Classe (0: sem diabetes, 1: com diabetes)')
plt.ylabel('Frequência')
plt.title('Distribuição das Classes')
plt.show()

# NÃO é necessário inverter classes (1 já é positivo para diabetes)

# Dividir em treino/teste (test_size=1/3, ~256 treino, ~512 teste; stratify para manter proporção ~35% positivos)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, stratify=y, random_state=42)

print(f"Proporção positivos no treino: {np.sum(y_train) / len(y_train):.3f}")
print(f"Proporção positivos no teste: {np.sum(y_test) / len(y_test):.3f}")

# 1. CLASSIFICADOR LogisticRegression SEM NORMALIZAÇÃO
lr_raw = LogisticRegression(max_iter=5000, random_state=42).fit(X_train, y_train)

# Coeficientes do modelo (pesos para cada feature)
w_raw = lr_raw.coef_.squeeze()
plt.figure(figsize=(8, 4))
plt.plot(w_raw, 'o-', markersize=4)
plt.xlabel('Índice da Feature')
plt.ylabel('Coeficiente')
plt.title('Coeficientes do LogisticRegression (sem normalização)')
plt.xticks(range(len(D['feature_names'])), D['feature_names'], rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Previsões no teste
y_test_pred_raw = lr_raw.predict(X_test)
print("Matriz de Confusão (sem normalização, threshold=0):")
print(confusion_matrix(y_test, y_test_pred_raw))
print(f"Erros totais: {np.sum(y_test != y_test_pred_raw)}")
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_test_pred_raw))

# Função de decisão (scores antes do threshold)
y_test_dec_raw = lr_raw.decision_function(X_test)

# Testar diferentes thresholds na função de decisão
thresholds = [0, 0.5, 1, 1.5, 2, -0.5, -1, -1.5, -2]
print("\nMatrizes de Confusão para diferentes thresholds (sem normalização):")
for thresh in thresholds:
    cm = confusion_matrix(y_test, y_test_dec_raw >= thresh)
    print(f"Threshold >= {thresh}: {cm}")

# Curva ROC
fpr_raw, tpr_raw, _ = roc_curve(y_test, y_test_dec_raw)
plt.figure(figsize=(6, 6))
plt.plot(fpr_raw, tpr_raw, label='ROC (sem normalização)')
plt.plot([0, 1], [0, 1], 'k--', label='Classificador Aleatório')
plt.xlabel('FPR (False Positive Rate)')
plt.ylabel('TPR (True Positive Rate)')
plt.title('Curva ROC - LogisticRegression (sem normalização)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('scaled')
plt.show()

# 2. COM NORMALIZAÇÃO (média 0, variância 1 por feature, como sugerido no enunciado)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

lr_scaled = LogisticRegression(max_iter=5000, random_state=42).fit(X_train_scaled, y_train)

# Coeficientes (agora nas features normalizadas)
w_scaled = lr_scaled.coef_.squeeze()
plt.figure(figsize=(8, 4))
plt.plot(w_scaled, 'o-', markersize=4)
plt.xlabel('Índice da Feature')
plt.ylabel('Coeficiente')
plt.title('Coeficientes do LogisticRegression (com normalização)')
plt.xticks(range(len(D['feature_names'])), D['feature_names'], rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Previsões no teste
y_test_pred_scaled = lr_scaled.predict(X_test_scaled)
print("\nMatriz de Confusão (com normalização, threshold=0):")
print(confusion_matrix(y_test, y_test_pred_scaled))
print(f"Erros totais: {np.sum(y_test != y_test_pred_scaled)}")
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_test_pred_scaled))

# Função de decisão
y_test_dec_scaled = lr_scaled.decision_function(X_test_scaled)

# Testar thresholds
print("\nMatrizes de Confusão para diferentes thresholds (com normalização):")
for thresh in thresholds:
    cm = confusion_matrix(y_test, y_test_dec_scaled >= thresh)
    print(f"Threshold >= {thresh}: {cm}")

# Curva ROC
fpr_scaled, tpr_scaled, _ = roc_curve(y_test, y_test_dec_scaled)
plt.figure(figsize=(6, 6))
plt.plot(fpr_raw, tpr_raw, label='Sem normalização')
plt.plot(fpr_scaled, tpr_scaled, label='Com normalização')
plt.plot([0, 1], [0, 1], 'k--', label='Classificador Aleatório')
plt.xlabel('FPR (False Positive Rate)')
plt.ylabel('TPR (True Positive Rate)')
plt.title('Curva ROC - Comparação Normalização vs. Raw')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('scaled')
plt.show()

# Análise rápida: Compare as métricas (ex: AUC pode ser calculada com roc_auc_score se importado)
from sklearn.metrics import roc_auc_score
print(f"AUC sem normalização: {roc_auc_score(y_test, y_test_dec_raw):.3f}")
print(f"AUC com normalização: {roc_auc_score(y_test, y_test_dec_scaled):.3f}")

# Próximos passos (para completar o trabalho):
# - Adicione outros 2 classificadores (ex: SVM, KNN) + RandomForestClassifier obrigatório
# - Use cross-validation para escolha de hiperparâmetros (GridSearchCV)
# - Compare todos com tabelas de métricas (accuracy, precision, recall, F1, AUC)
# - Justifique: Normalização beneficia modelos sensíveis a escala (como LR); use CV para estimativa robusta
# - Salve plots e resultados em células markdown para análise
```
